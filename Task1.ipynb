```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
poly_data = pd.read_csv('rank_salary.csv')
```

```python
poly_data.head()
```

```python
X_poly = poly_data[['IndependentVariable']]
y_poly = poly_data['DependentVariable']
```

```python
poly_features = PolynomialFeatures(degree=4)
X_poly_transformed = poly_features.fit_transform(X_poly)
```

```python
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly_transformed, y_poly, test_size=0.2, random_state=0)
```

```python
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train_poly)
```

```python
y_pred_poly = poly_model.predict(poly_features.transform(X_test_poly))
```

```python
plt.scatter(X_poly, y_poly, color='red')
plt.scatter(X_poly, poly_model.predict(poly_features.fit_transform(X_poly)), color='blue')
plt.title('Polynomial Regression Results')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```
